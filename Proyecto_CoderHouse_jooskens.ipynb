```python
import pandas as pd
import zipfile
import io
import requests
import matplotlib.pyplot as plt
import seaborn as sns

# Ruta del archivo CSV en Google Drive
file_path = '/content/gdrive/My Drive/CoderEntregable/student-mat.csv'

df = pd.read_csv(file_path, sep=';')

# Mostrar  dataset
print(df.head())
```

```python
#Numero de estudiantes
plt.figure(figsize=(10, 6))
df['sex'].value_counts().sort_index().plot(kind='bar', grid=True)
plt.title('Número de estudiantes por género')
plt.xlabel('Género')
plt.ylabel('Número de estudiantes')
plt.show()
```

```python
#Horas de estudio y notas finales
plt.figure(figsize=(10, 6))
plt.scatter(df['studytime'], df['G3'], alpha=0.5)
plt.title('Horas de estudio vs. Notas finales')
plt.xlabel('Horas de estudio')
plt.ylabel('Nota final')
plt.grid(True)
plt.show()
```

```python
##gráficos con Seaborn##
##Numero de estudiantes por genero##
plt.figure(figsize=(10, 6))
df['pass'] = df['G3'] >= 10
sns.countplot(data=df, x='sex', hue='pass')
plt.title('Número de estudiantes por género y aprobación')
plt.xlabel('Género')
plt.ylabel('Número de estudiantes')
plt.grid(True)
plt.show()
```

```python
# Histograma de las notas finales con hue para diferenciar por género
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='G3', hue='sex', multiple='stack', bins=30, kde=False)
plt.title('Distribución de notas finales por género')
plt.xlabel('Nota final')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
## Distribución de notas finales por género##
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='sex', y='G3')
plt.title('Distribución de notas finales por género')
plt.xlabel('Género')
plt.ylabel('Nota final')
plt.grid(True)
plt.show()
```
